```python
import pandas as pd
import re
from pathlib import Path

files = [
    Path("./01_raw_data/subsidy/무공해차 통합누리집2020승용 화물.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2021승용.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2021화물.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2022승용.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2022화물.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2023승용.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2023화물.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2024승용.html"),
    Path("./01_raw_data/subsidy/무공해차 통합누리집2024화물.html"),
]

output_dir = Path("./02_interim_data/subsidy")
output_dir.mkdir(parents=True, exist_ok=True)

all_dfs = []

for file in files:
    tables = pd.read_html(file)
    df = tables[0]
    m = re.search(r"(20\d{2})", file.stem)
    year = int(m.group(1)) if m else pd.NA
    df["년도"] = year
    t = re.search(r"(승용|화물)", file.stem)
    df["차종"] = t.group(1) if t else pd.NA
    all_dfs.append(df)

merged_df = pd.concat(all_dfs, ignore_index=True)
merged_df.drop(columns=['모델별 지방비 조회'], 
               errors='ignore', inplace=True)

merged_df

```

```python

col_a = "보조금/승용(만원) (국비+지방비)"
col_b = "보조금/승용(만원) (국비+지방비+보급목표이행보조금등)"

merged_df["보조금/승용(만원)"] = merged_df.get(col_a).combine_first(merged_df.get(col_b))
merged_df["보조금/승용(만원)"] = (
    merged_df["보조금/승용(만원)"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.extract(r"(\d+(?:\.\d+)?)")[0]
    .astype(float)
)
for c in (col_a, col_b):
    if c in merged_df.columns:
        merged_df.drop(columns=c, inplace=True)

cargo_col = "보조금/화물(만원) (국비+지방비)"
classes = ["초소형", "경형", "소형"]

def parse_cargo(s: str):
    out = {k: np.nan for k in classes}
    if s is None or (isinstance(s, float) and np.isnan(s)):
        return out
    text = str(s)
    for m in re.finditer(r"(초소형|경형|소형)\s*([\d,]+)", text):
        k = m.group(1)
        v = float(m.group(2).replace(",", ""))
        out[k] = v

    return out

cargo_expanded = merged_df[cargo_col].apply(parse_cargo).apply(pd.Series)

cargo_expanded = cargo_expanded.rename(columns={k: f"화물/{k}(만원)" for k in cargo_expanded.columns})

merged_df = pd.concat([merged_df.drop(columns=[cargo_col], errors="ignore"), cargo_expanded], axis=1)

merged_df.drop(['보조금/초소형(만원) (국비+지방비)', '보조금/승합(만원) (국비+지방비)'], axis=1, inplace=True)
merged_df= merged_df[~merged_df["시도"].isin(["협회", "공단"])]
merged_df
```

```python

avg_df = (
    merged_df
    .groupby(["년도","시도"], as_index=False)[
        ["보조금/승용(만원)", "화물/초소형(만원)", "화물/경형(만원)", "화물/소형(만원)"]
    ]
    .mean(numeric_only=True)
)
nationwide_df = (
    merged_df
    .groupby(["년도"], as_index=False)[
        ["보조금/승용(만원)", "화물/초소형(만원)", "화물/경형(만원)", "화물/소형(만원)"]
    ]
    .mean(numeric_only=True)
)
nationwide_df["시도"] = "전국"
nationwide_df = nationwide_df[avg_df.columns]

avg_df = pd.concat([avg_df, nationwide_df], ignore_index=True)
avg_df = avg_df.fillna(0)
avg_df.rename(columns={"시도": "지역"}, inplace=True)
# avg_df.to_csv('./02_interim_data/subsidy.csv', index=False, encoding='utf-8-sig')
avg_df
```

```python

passenger = (
    avg_df[["년도", "지역", "보조금/승용(만원)"]]
      .assign(차종="승용",
              **{"초소형(만원)": np.nan, "경형(만원)": np.nan, "소형(만원)": np.nan})
      .rename(columns={"보조금/승용(만원)": "승용(만원)"})
)

cargo = (
    avg_df[["년도", "지역", "화물/초소형(만원)", "화물/경형(만원)", "화물/소형(만원)"]]
      .assign(차종="화물", **{"승용(만원)": np.nan})
      .rename(columns={
          "화물/초소형(만원)": "초소형(만원)",
          "화물/경형(만원)": "경형(만원)",
          "화물/소형(만원)": "소형(만원)",
      })
)

cols = ["년도", "지역", "차종", "승용(만원)", "초소형(만원)", "경형(만원)", "소형(만원)"]
out = (
    pd.concat([passenger[cols], cargo[cols]], ignore_index=True)
      .sort_values(["년도", "지역", "차종"], ignore_index=True)
)
out = out.rename(columns={
    '승용(만원)':'승용차 보조금',
    '초소형(만원)':'초소형 화물차 보조금',
    '경형(만원)':'경형 화물차 보조금',
    '소형(만원)':'소형 화물차 보조금'
      })

out.head(50)
```

```python
out2 = out.copy()

out2["보조금"] = np.nan
out2.loc[out2["차종"] == "승용", "보조금"] = out2.loc[out2["차종"] == "승용", "승용차 보조금"]
out2.loc[out2["차종"] == "화물", "보조금"] = out2.loc[out2["차종"] == "화물", "소형 화물차 보조금"]

out2 = out2[["년도", "지역", "차종", "보조금"]]
out2.to_csv("./02_interim_data/subsidy.csv", index=False, encoding="utf-8-sig")
out2.head(50)
```
