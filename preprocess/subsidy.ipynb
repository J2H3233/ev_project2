```python
import os
import re
import glob
import pandas as pd
from bs4 import BeautifulSoup

# 원본 HTML이 있는 폴더 (예: "./01_raw_data/subsidy")
BASE_DIR = "./01_raw_data/subsidy"

# 대상 파일 패턴
patterns = [
    os.path.join(BASE_DIR, "무공해차 통합누리집*.html"),
    os.path.join(BASE_DIR, "무공해차 통합누리집승용*.html"),
]

def pick_amount_cell(cols):
    """
    행의 <td>들 중에 '초소형/경형/소형' 키워드를 포함한 셀을 찾아 반환.
    없으면 마지막 셀을 반환(사이트 버전별 컬럼 위치 변동 대응).
    """
    for td in cols:
        txt = td.get_text(" ", strip=True)
        if any(k in txt for k in ("초소형", "경형", "소형")):
            return td
    return cols[-1]

def parse_amounts(text):
    """
    '... 초소형 400, 경형 500, 소형 600 ...' 같은 텍스트에서 금액만 추출.
    """
    def to_int(m):
        return int(m.group(1).replace(",", "")) if m else None

    small  = to_int(re.search(r"초\s*소\s*형\s*([0-9,]+)", text)) or to_int(re.search(r"초소형\s*([0-9,]+)", text))
    light  = to_int(re.search(r"경\s*형\s*([0-9,]+)",   text)) or to_int(re.search(r"경형\s*([0-9,]+)",   text))
    medium = to_int(re.search(r"소\s*형\s*([0-9,]+)",   text)) or to_int(re.search(r"소형\s*([0-9,]+)",   text))
    return small, light, medium

def parse_meta_from_filename(fname):
    year = re.search(r"(20\d{2})", fname)
    year = year.group(1) if year else None
    car_type = "승용" if "승용" in fname else ("화물" if "화물" in fname else None)
    return year, car_type

rows = []

for pattern in patterns:
    for path in glob.glob(pattern):
        year, car_type = parse_meta_from_filename(os.path.basename(path))

        with open(path, "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")

        # 표 행 선택: 기본은 table.table01 tbody tr, 없으면 모든 tr fallback
        trs = soup.select("table.table01 tbody tr")
        if not trs:
            trs = soup.select("table tbody tr")
        # 헤더/빈행 제거
        trs = [tr for tr in trs if tr.find_all("td")]

        for tr in trs:
            cols = tr.find_all("td")
            if len(cols) < 2:
                continue

            sido   = cols[0].get_text(strip=True)
            region = cols[1].get_text(strip=True)

            # 금액 들어있는 셀 찾기
            target_td = pick_amount_cell(cols)
            info_text = target_td.get_text(separator="\n", strip=True)

            small, light, medium = parse_amounts(info_text)

            rows.append([
                year, car_type, sido, region, small, light, medium, os.path.basename(path)
            ])

# DataFrame
df = pd.DataFrame(
    rows,
    columns=["년도", "차종", "시도", "지역구분", "초소형(만원)", "경형(만원)", "소형(만원)", "파일명"]
)

# 숫자 컬럼 정리(전부 int로 캐스팅 가능하면 캐스팅)
for col in ["초소형(만원)", "경형(만원)", "소형(만원)"]:
    df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")

# 필요시 저장
# out_path = os.path.join(BASE_DIR, "무공해차_보조금정리.csv")
# df.to_csv(out_path, index=False, encoding="utf-8-sig")
# print("저장:", out_path)

df.head(50)
```

```python
import os
import re
from bs4 import BeautifulSoup
import pandas as pd

# HTML 파일들이 저장된 폴더 경로
folder_path = "./01_raw_data/subsidy"

# 결과 저장 리스트
data = []

# 파일명에서 연도와 차종 추출하는 함수
def parse_filename(filename):
    year_match = re.search(r"20\d{2}", filename)
    year = int(year_match.group()) if year_match else None
    if "승용" in filename:
        type_ = "승용"
    elif "화물" in filename:
        type_ = "화물"
    else:
        type_ = "혼합"
    return year, type_

# 지역, 차량 분류 추출 함수
def extract_info_from_html(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    rows = soup.find_all("tr")

    for row in rows:
        cols = [c.get_text(strip=True) for c in row.find_all(["td", "th"])]
        if not cols:
            continue

        # 지역명은 보통 첫 번째 컬럼
        region = cols[0]

        # 초소형/경형/소형 키워드 탐색
        category = None
        for keyword in ["초소형", "경형", "소형"]:
            if any(keyword in col for col in cols):
                category = keyword
                break

        if category:
            yield region, category

# 모든 HTML 파일 순회
for filename in os.listdir(folder_path):
    if filename.endswith(".html") and filename.startswith("무공해차"):
        year, car_type = parse_filename(filename)
        with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
            html_content = f.read()

        for region, category in extract_info_from_html(html_content):
            data.append({
                "년도": year,
                "차종": car_type,
                "지역": region,
                "차량 분류": category
            })

# DataFrame으로 변환
df = pd.DataFrame(data)

# 중복 제거 및 정렬
df = df.drop_duplicates().sort_values(["년도", "차종", "지역", "차량 분류"])

# # 결과 저장
# output_path = "/mnt/data/무공해차_정리.xlsx"
df

```

```python
import os
import re
import pandas as pd
from bs4 import BeautifulSoup

# HTML 파일이 있는 폴더 경로
folder_path = "./01_raw_data/subsidy"

# 결과를 담을 리스트
data = []

# 파일명에서 년도와 차종(승용/화물) 추출
def parse_filename(filename):
    year_match = re.search(r"20\d{2}", filename)
    year = year_match.group() if year_match else None
    car_type = "승용" if "승용" in filename else "화물"
    return year, car_type

# 폴더 내 HTML 파일 읽기
for file_name in os.listdir(folder_path):
    if file_name.endswith(".html"):
        year, car_type = parse_filename(file_name)

        with open(os.path.join(folder_path, file_name), "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")

        # 표(tag) 찾기
        tables = soup.find_all("table")
        for table in tables:
            headers = [th.get_text(strip=True) for th in table.find_all("th")]
            if any(k in headers for k in ["지역", "초소형", "경형", "소형"]):
                for row in table.find_all("tr")[1:]:
                    cells = [td.get_text(strip=True) for td in row.find_all("td")]
                    if len(cells) >= 4:
                        region = cells[0]
                        small = cells[1] if len(cells) > 1 else ""
                        light = cells[2] if len(cells) > 2 else ""
                        compact = cells[3] if len(cells) > 3 else ""
                        data.append([year, car_type, region, small, light, compact])

# DataFrame 생성
df = pd.DataFrame(data, columns=["년도", "차종", "지역", "초소형", "경형", "소형"])

# # CSV 저장
# output_path = "/mnt/data/무공해차_정리.csv"
data
# df.to_csv(output_path, index=False, encoding="utf-8-sig")

# print(f"저장 완료: {output_path}")
# df.head()
```
